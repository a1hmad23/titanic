# functions

```python
def make_col_off_second_word(df, old_col, new_col):
    df[new_col] = df[old_col].str.split(' ').apply(lambda x: x[1])
    return df
```

```python
def group_title(df, col):
    df[col] = df[col].replace({
    'Capt.':'Military', 
    'Major.':'Military',
    'Col.':'Military',
    'Jonkheer.':'Noble',
    'the Countess':'Noble',
    'Don.':'Noble',
    'Lady':'Noble',
    'Sir':'Noble',
    'Mlle.':'Noble',
    'Ms.':'Noble',
    'Mme.':'Noble'
})
    return df
```

```python
def zero_or_greater(df, col):
    df['binary_' + col] = df[col].map(lambda x: 0 if x == 0 else 1)
    return df
```

```python
def col_per_counts_col(df, col, counts_col):
    counts = df[counts_col].map(df[counts_col].value_counts())
    df[col + ' per ' + counts_col + ' count'] = df[col] / no_of_Tickets
    return df
```

```python
def make_col_off_rep_values(df, old_col):
    df['Shared_' + old_col] = df[old_col].map(df[old_col].value_counts() > 1)
    return df
```

```python
def make_col_off_first_letter(df, old_col, new_col):
    df[new_col] = df[old_col].str[0]
    return df
```

```python
from sklearn.preprocessing import FunctionTransformer
from sklearn.pipeline import Pipeline

title_pipeline = Pipeline([
    ('make_title', FunctionTransformer(make_col_off_second_word, kw_args={'old_col':'Name', 'new_col':'Title'}, validate=False, feature_names_out="one-to-one")),
    ('group_title', FunctionTransformer(group_title, kw_args={'col':'Title'}, validate=False, feature_names_out='one-to-one'))
])
```

```python
fare_pipeline = Pipeline([
    ('fare_per_person', FunctionTransformer(col_per_counts_col, kw_args={'col':'Fare', 'counts_col':'Ticket'}, validate=False, feature_names_out='one-to-one')),
    ('fare_cut', FunctionTransformer(pd.cut, kw_args={'bins':fare_bins, 'labels':'fare_labels'}, validate=False, feature_names_out='one-to-one'))
])
```

```python
from sklearn.compose import ColumnTransformer

features_ct = ColumnTransformer(
    transformers=[
        ('title_pipeline', title_pipeline, ['Name']),
        ('make_deck', FunctionTransformer(make_col_off_first_letter, kw_args={'old_col':'Cabin', 'new_col':'Deck'}, validate=False, feature_names_out='one-to-one'), ['Cabin']),
        ('shared_cabin', FunctionTransformer(make_col_off_rep_values, kw_args={'old_col':'Cabin'}, validate=False, feature_names_out='one-to-one'), ['Cabin']),
        ('shared_ticket', FunctionTransformer(make_col_off_rep_values, kw_args={'old_col':'Ticket'}, validate=False, feature_names_out='one-to-one'), ['Ticket']),
        ('fare_pipeline', fare_pipeline, ['Fare', 'Ticket']),
        ('bool_sib', FunctionTransformer(zero_or_greater, kw_args={'col':'SibSp'}, validate=False, feature_names_out='one-to-one'), ['SibSp']),
        ('bool_par', FunctionTransformer(zero_or_greater, kw_args={'col':'Parch'}, validate=False, feature_names_out='one-to-one'), ['Parch'])
    ],
    remainder='passthrough',
    verbose_feature_names_out=False
).set_output(transform="pandas")
```
